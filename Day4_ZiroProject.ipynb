```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
connection = sqlite3.connect("RecruitSmart.db")

cursor = connection.cursor()

print("Database Created Successfully!")
```

```python
jobs = pd.read_csv(r"C:\Users\cbala\OneDrive\Desktop\RecruitSmart\data\jobs_cleaned.csv")

people = pd.read_csv(r"C:\Users\cbala\OneDrive\Desktop\RecruitSmart\data\01_people.csv")

applications = pd.read_csv(r"C:\Users\cbala\OneDrive\Desktop\RecruitSmart\data\applications.csv")

person_skills = pd.read_csv(r"C:\Users\cbala\OneDrive\Desktop\RecruitSmart\data\05_person_skills.csv")

skills = pd.read_csv(r"C:\Users\cbala\OneDrive\Desktop\RecruitSmart\data\06_skills.csv")
```

```python
jobs.to_sql("Jobs", connection, if_exists="replace", index=False)

people.to_sql("Candidates", connection, if_exists="replace", index=False)

applications.to_sql("Applications", connection, if_exists="replace", index=False)

person_skills.to_sql("PersonSkills", connection, if_exists="replace", index=False)

skills.to_sql("Skills", connection, if_exists="replace", index=False)

print("Tables Imported Successfully!")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, connection)
```

```python
query = """
SELECT *
FROM Jobs
LIMIT 5;
"""

pd.read_sql(query, connection)
```

```python
cursor.execute("""

CREATE VIEW IF NOT EXISTS JobCandidateView AS

SELECT

a.application_id,

j.job_id,

j.title AS job_title,

p.person_id,

p.name

FROM Applications a

JOIN Jobs j
ON a.job_id=j.job_id

JOIN Candidates p
ON a.person_id=p.person_id;

""")

connection.commit()

print("SQL View Created Successfully!")
```

```python
query = """
SELECT *
FROM JobCandidateView
LIMIT 10;
"""

pd.read_sql(query, connection)
```

```python
connection.close()

print("Database Closed")
```
