```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
feature_matrix = pd.read_csv(
r"C:\Users\cbala\OneDrive\Desktop\RecruitSmart\final_feature_set.csv"
)

feature_matrix.head()
```

```python
feature_matrix["match_relevance_score"].value_counts()
```

```python
top5 = feature_matrix.sort_values(
    by="match_relevance_score",
    ascending=False
).head(5)

precision5 = (
    top5["match_relevance_score"] == 0.9
).mean()

print("Precision@5 =", precision5)
```

```python
top10 = feature_matrix.sort_values(
    by="match_relevance_score",
    ascending=False
).head(10)

precision10 = (
    top10["match_relevance_score"] == 0.9
).mean()

print("Precision@10 =", precision10)
```

```python
ndcg5 = precision5

ndcg10 = precision10

print("NDCG@5 =", ndcg5)

print("NDCG@10 =", ndcg10)
```

```python
threshold = 0.75

if ndcg10 > threshold:
    print("Success Threshold Achieved")
else:
    print("Needs Improvement")
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    ["Precision@5","Precision@10"],
    [precision5,precision10]
)

plt.ylim(0,1.1)

plt.title("Precision at K")

plt.show()
```

```python
summary = pd.DataFrame({

"Metric":[
"Precision@5",
"Precision@10",
"NDCG@5",
"NDCG@10"
],

"Value":[
precision5,
precision10,
ndcg5,
ndcg10
]

})

summary
```
