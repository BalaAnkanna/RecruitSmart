```python
from fastapi import FastAPI
from pydantic import BaseModel
```

```python
app = FastAPI()

print("FastAPI App Created Successfully!")
```

```python
class Resume(BaseModel):
    person_id: int
    skills: str
    experience: str
```

```python
class ResumePreprocessor:

    def preprocess(self, resume):

        return {
            "person_id": resume.person_id,
            "skills": resume.skills.lower(),
            "experience": resume.experience.lower()
        }
```

```python
sample = Resume(
    person_id=1,
    skills="Python SQL Machine Learning",
    experience="Entry Level"
)

processor = ResumePreprocessor()

processor.preprocess(sample)
```

```python
@app.post("/predict")
def predict(resume: Resume):

    processor = ResumePreprocessor()

    processed = processor.preprocess(resume)

    return {
        "Match Score":0.91,
        "Recommended Job":"Data Analyst",
        "Processed Resume":processed
    }
```

```python
predict(sample)
```
