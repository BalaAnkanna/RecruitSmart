```python
import re
```

```python
def clean_text(text):

    if pd.isnull(text):
        return ""

    text = text.lower()

    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)

    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
jobs["clean_description"] = jobs["description"].apply(clean_text)

jobs["clean_requirements"] = jobs["requirements"].apply(clean_text)
```
