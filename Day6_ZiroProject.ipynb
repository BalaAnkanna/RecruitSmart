```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect("RecruitSmart.db")
```

```python
query = """

SELECT
job_id,
required_experience,

CASE
WHEN required_experience='Internship' THEN 0
WHEN required_experience='Entry level' THEN 1
WHEN required_experience='Associate' THEN 2
WHEN required_experience='Mid-Senior level' THEN 3
WHEN required_experience='Executive' THEN 4
ELSE NULL
END AS experience_score,

ROUND(
(
CASE
WHEN required_experience='Internship' THEN 0
WHEN required_experience='Entry level' THEN 1
WHEN required_experience='Associate' THEN 2
WHEN required_experience='Mid-Senior level' THEN 3
WHEN required_experience='Executive' THEN 4
ELSE NULL
END
)/4.0,2) AS normalized_score

FROM Jobs

LIMIT 20;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT
job_id,
title,
requirements

FROM Jobs

WHERE

requirements LIKE '%AWS%'

OR requirements LIKE '%Azure%'

OR requirements LIKE '%Cisco%'

OR requirements LIKE '%Oracle%'

LIMIT 10;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT
job_id,
title,
requirements

FROM Jobs

WHERE

requirements LIKE '%Python%'

OR requirements LIKE '%SQL%'

OR requirements LIKE '%Java%'

OR requirements LIKE '%Excel%'

OR requirements LIKE '%Machine Learning%'

LIMIT 10;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT

job_id,

COUNT(*) AS frequency

FROM Jobs

GROUP BY job_id

ORDER BY frequency DESC;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT

title,

COUNT(*) AS total_jobs

FROM Jobs

GROUP BY title

ORDER BY total_jobs DESC

LIMIT 20;

"""

pd.read_sql(query, connection)
```

```python
connection.close()

print("Database Closed Successfully!")
```
