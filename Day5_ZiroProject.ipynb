```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect("RecruitSmart.db")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, connection)
```

```python
pd.read_sql("SELECT * FROM Skills LIMIT 10;", connection)
```

```python
pd.read_sql("SELECT * FROM PersonSkills LIMIT 10;", connection)
```

```python
query = """

SELECT
    person_id,
    COUNT(skill) AS total_skills

FROM PersonSkills

GROUP BY person_id

ORDER BY total_skills DESC;

"""

pd.read_sql(query, connection)
```

```python
pd.read_sql("""
SELECT
    person_id,
    COUNT(*) AS duplicate_count
FROM Candidates
GROUP BY person_id
HAVING COUNT(*) > 1;
""", connection)
```

```python
pd.read_sql("""
SELECT
    application_id,
    COUNT(*) AS duplicate_count
FROM Applications
GROUP BY application_id
HAVING COUNT(*) > 1;
""", connection)
```

```python
query = """
SELECT
job_id,

required_experience,

CASE

WHEN required_experience='Internship' THEN 0

WHEN required_experience='Entry level' THEN 1

WHEN required_experience='Associate' THEN 2

WHEN required_experience='Mid-Senior level' THEN 3

WHEN required_experience='Executive' THEN 4

ELSE NULL

END AS experience_score

FROM Jobs

LIMIT 20;
"""

pd.read_sql(query, connection)
```
